# Aufgabe 3 - Strahlungstemperatur

```python
from math import *
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
# Aufgabe 3

# Lichtgeschwindigkeit
c = 3.0e+8  # [m/s]
v = c/np.linspace(100.0e-9,1.0e-2,100000) # Frequenz [Hz]
h = 6.626e-34   # Planck Konstane [J/s]
T = 255 # [K]
k = 1.38e-23   # Boltzmann Konstante [J/K]
sig = 5.67e-8  # Boltzmann Konstante [Wm-2K-1]

# Rayleigh-Jeans
Lb = (2*k*T*(v**2)) /c**2 # all values

# Wien
Lb2 = 2.0*h*(v**3)/(c**2 * np.exp((h*v)/(k*T)) )
    
# Planck (Frequenz)
def Lv(x,T):
    a = (2*h*x**3 /c**2)
    b = (h*x)/(k*T)
    return a / ( np.exp(b) - 1)


Lv1 = Lv(v,T)

# Plot
f,ax = plt.subplots(2,1,figsize=(16,9))
ax[0].plot(v,Lv1,label='Planck')
ax[0].plot(v,Lb,label='Rayleigh-Jeans')
ax[0].plot(v,Lb2,label='Wien')
ax[0].set_xscale('log')
ax[0].set_title('spektrale Radianz')
ax[0].set_ylim(0,4e-12)
ax[0].set_ylabel(r'$L_\nu$ [$Wm^{-2}$]')
ax[0].legend()
ax[0].grid()
ax[1].plot(v,Lb/Lv1,'g',label='RJ/P')
ax[1].plot(v,Lb2/Lv1,'r',label='W/P')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_xlabel(r'$\nu\ [Hz]$')
ax[1].set_ylabel(r'Quotient [1]')
ax[1].set_ylim(-1e-4,1e4)
ax[1].grid()
ax[1].legend()
```
